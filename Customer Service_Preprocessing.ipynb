## Getting dummies for the Service_Type_Code

```python
service_columns = pd.get_dummies(data['Service_Type_Code'])
```

```python
service_columns
```

```python
service_columns['check'] = service_columns.sum(axis=1)
service_columns
```

```python
service_columns['check'].sum()
```

```python
service_columns = service_columns.drop(['check'], axis = 1)
service_columns
```

# Group the Service Type Codes:

```python
data.columns.values
```

```python
service_columns.columns.values
```

```python
data = data.drop(['Service_Type_Code'], axis = 1)
```

```python
data.head()
```

```python
service_type_1 = service_columns.loc[:, 1:8].max(axis=1)
service_type_2 = service_columns.loc[:, 9:15].max(axis=1)
service_type_3 = service_columns.loc[:, 16:23].max(axis=1)
service_type_4 = service_columns.loc[:, 24:].max(axis=1)
```

```python
#Concatenate Column Values
data = pd.concat([data, service_type_1, service_type_2, service_type_3, service_type_4], axis = 1)
data.head()
```

```python
data.columns.values
```

```python
column_names=['Request_Date', 'Service_Cost', 'Distance_to_Service_Center',
       'Customer_Age', 'Daily_Request_Volume_Avg',
       'Customer_Satisfaction_Index', 'Customer_Tier',
       'Previous_Requests', 'Support_Contacts',
       'Issue_Resolution_Time_Hours','service_type_1', 'service_type_2', 'service_type_3', 'service_type_4']
```

```python
data.columns = column_names
```

```python
data.head()
```

```python
data.columns
```

```python
column_names_reordered=['service_type_1',
       'service_type_2', 'service_type_3', 'service_type_4', 'Request_Date', 'Service_Cost', 'Distance_to_Service_Center',
       'Customer_Age', 'Daily_Request_Volume_Avg',
       'Customer_Satisfaction_Index', 'Customer_Tier', 'Previous_Requests',
       'Support_Contacts', 'Issue_Resolution_Time_Hours']
```

```python
data = data[column_names_reordered]
```

```python
data.head()
```

```python
data_service_mod = data.copy()
```

```python
data_service_mod.head()
```

```python
# Working on the Request_Date column
```

```python
type(data_service_mod['Request_Date'][0])
```

```python
data_service_mod['Request_Date'] = pd.to_datetime(data_service_mod['Request_Date'], format = '%d/%m/%Y')
```

```python
type(data_service_mod['Request_Date'][0])
```

```python
data_service_mod.info()
```

```python
data_service_mod['Request_Date'][0].month
```

```python
list_months = []
list_months
```

```python
data_service_mod.shape
```

```python
for i in range(data_service_mod.shape[0]):
    list_months.append(data_service_mod['Request_Date'][i].month)
```

```python
len(list_months)
```

```python
data_service_mod['Month Value'] = list_months
```

```python
data_service_mod.head()
```

```python
# Extracting the day of the week
```

```python
data_service_mod['Request_Date'][699].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
data_service_mod['Day of the Week'] = data_service_mod['Request_Date'].apply(date_to_weekday)
```

```python
data_service_mod.head()
```

```python
data_service_mod = data_service_mod.drop(['Request_Date'], axis = 1)
data_service_mod.head()
```

```python
data_service_mod.columns
```

```python
column_names_upd=['service_type_1', 'service_type_2', 'service_type_3', 'service_type_4', 'Month Value', 'Day of the Week',
       'Service_Cost', 'Distance_to_Service_Center', 'Customer_Age',
       'Daily_Request_Volume_Avg', 'Customer_Satisfaction_Index',
       'Customer_Tier', 'Previous_Requests', 'Support_Contacts',
       'Issue_Resolution_Time_Hours']
```

```python
data_service_mod = data_service_mod[column_names_upd]
data_service_mod.head()
```

```python
data_preprocessed = data_service_mod.copy()
data_preprocessed.head(10)
```
